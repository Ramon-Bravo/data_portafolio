### [B1] Diferencia entre miércoles y sábados para  `'order_hour_of_day'`. Traza gráficos de barra para los dos días y describe las diferencias que veas.

```python
miercoles_hr = df_ordenes[df_ordenes['order_dow'] == 3]['order_hour_of_day'].value_counts().reset_index()
miercoles_hr = miercoles_hr.rename(columns={'index': 'Horas', 'order_hour_of_day': 'Pedidos'})
```

```python
sabado_hr = df_ordenes[df_ordenes['order_dow'] == 6]['order_hour_of_day'].value_counts().reset_index()
sabado_hr = sabado_hr.rename(columns={'index': 'Horas', 'order_hour_of_day': 'Pedidos'}) 
```

```python
df_juntos = miercoles_hr.merge(
    sabado_hr,
    on='Horas',
    sort=True,
    suffixes=['_miercoles', '_sabado']
)
df_juntos
```

```python
df_juntos.plot(
    kind='bar',
    title='Comparativa de ordenes por hora Miércoles vs. Sábado.',
    x='Horas',
    y=['Pedidos_miercoles', 'Pedidos_sabado'],
    xlabel='Hora del día',
    ylabel='Ordenes',
    figsize=(6,6)
)
plt.legend=(['Miércoles', 'Sábados'])
plt.show()
```
